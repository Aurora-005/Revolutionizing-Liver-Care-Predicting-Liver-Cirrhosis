```python
import torch

from torch.utils.data import Dataset

class CustomDataset(Dataset):

  def __init__(self,texts,labels,tokenizer,max_len):

    self.texts=texts

    self.labels=labels

    self.tokenizer=tokenizer

    self.max_len=max_len

  def __len__(self):

    return len(self.texts)

  def __getitem__(self,idx):

    text=self.texts[idx]

    label=self.labels[idx]

    encoding=self.tokenizer.encode_plus(

        text,

        add_special_tokens=True,

        max_length=self.max_len,

        truncation=True,

        padding="max_length",

        return_attention_mask=True,

        return_tensors="pt"

    )

    return{

        "input_ids":encoding["input_ids"].flatten(),

        "attention_mask":encoding["attention_mask"].flatten(),

        "label":torch.tensor(label,dtype=torch.long)

    }


```

```python
tokenizer = get_tokenizer("basic_english")

# Define a generator function `yield_tokens` to yield tokens from the data iterator.
# The data iterator provides pairs of (label, text) where text is the input sentence.
def yield_tokens(data_iter):
    for _, text in data_iter:
        yield tokenizer(text)

# Build the vocabulary from the tokens yielded by the `yield_tokens` generator function.
# We set `min_freq=2` to include a token only if it appears more than 2 times in the dataset.

# We will also add "special" tokens that we'll use to signal something to our model
# <pad> is a padding token that is added to the end of a sentence to ensure
# the length of all sequences in a batch is the same
# <sos> signals the "Start-Of-Sentence" aka the start of the sequence
# <eos> signal the "End-Of-Sentence" aka the end of the sequence
# <unk> "unknown" token is used if a token is not contained in the vocab
vocab = build_vocab_from_iterator(
    yield_tokens(dataset_train),  # Tokenized data iterator
    min_freq=2,  # Minimum frequency threshold for token inclusion
    specials=['<pad>', '<sos>', '<eos>', '<unk>'],  # Special case tokens
    special_first=True  # Place special tokens first in the vocabulary
)

# Set the default index of the vocabulary to the index of the '<unk>' token.
# If a token is not found in the vocabulary, it will be replaced with the '<unk>' token.
vocab.set_default_index(vocab['<unk>'])
```

```python
text_tranform = T.Sequential(
    # Convert the sentences to indices based on the given vocabulary
    T.VocabTransform(vocab=vocab),
    # Add <sos> at the beginning of each sentence. 1 is used because the index for <sos> in the vocabulary is 1.
    T.AddToken(1, begin=True),
    # Crop the sentence if it is longer than the max length
    T.Truncate(max_seq_len=max_len),
    # Add <eos> at the end of each sentence. 2 is used because the index for <eos> in the vocabulary is 2.
    T.AddToken(2, begin=False),
    # Convert the list of lists to a tensor. This also pads a sentence with the <pad> token if it is shorter than the max length,
    # ensuring that all sentences are the same length.
    T.ToTensor(padding_value=0)
)
```

```python
import torch.nn as nn
class LSTM(nn.Module):
    def __init__(self, num_emb, output_size, num_layers=1, hidden_size=128):
        super(LSTM, self).__init__()

        # Create an embedding layer to convert token indices to dense vectors
        self.embedding = nn.Embedding(num_emb, hidden_size)

        # Define the LSTM layer
        self.lstm = nn.LSTM(input_size=hidden_size, hidden_size=hidden_size,
                            num_layers=num_layers, batch_first=True, dropout=0.5)

        # Define the output fully connected layer
        self.fc_out = nn.Linear(hidden_size, output_size)

    def forward(self, input_seq, hidden_in, mem_in):
        # Convert token indices to dense vectors
        input_embs = self.embedding(input_seq)

        # Pass the embeddings through the LSTM layer
        output, (hidden_out, mem_out) = self.lstm(input_embs, (hidden_in, mem_in))

        # Pass the LSTM output through the fully connected layer to get the final output
        return self.fc_out(output), hidden_out, mem_out
```

```python
# Set the device to GPU if available, otherwise fallback to CPU
device = torch.device(0 if torch.cuda.is_available() else 'cpu')

# Define the size of the hidden layer and number of LSTM layers
hidden_size = 64
num_layers = 3

# Create the LSTM classifier model
lstm_classifier = LSTM(num_emb=len(vocab), output_size=4,
                       num_layers=num_layers, hidden_size=hidden_size).to(device)

# Initialize the optimizer with Adam optimizer
optimizer = optim.Adam(lstm_classifier.parameters(), lr=learning_rate)

# Define the loss function as CrossEntropyLoss for classification
loss_fn = nn.CrossEntropyLoss()

# Initialize lists to store training and test loss, as well as accuracy
training_loss_logger = []
test_loss_logger = []
training_acc_logger = []
test_acc_logger = []
```

```python
import torch

# Define a prediction function
def predict_category(text, model, text_tokenizer, text_transform, device):
    """
    Predicts the category of a given text input.

    Parameters:
    - text (str): The input text to classify.
    - model: The trained LSTM model.
    - text_tokenizer: The tokenizer function used to tokenize text.
    - text_transform: The transformation function to convert text into tensor format.
    - device: The device to use for computation (CPU or GPU).

    Returns:
    - category (int): The predicted category label for the input text.
    """
    # Set the model to evaluation mode
    model.eval()

    # Preprocess the text (tokenize and transform)
    text_tokens = text_transform(text_tokenizer([text])).to(device)

    # Initialize hidden and memory states
    bs = 1  # batch size of 1 for a single input
    hidden = torch.zeros(num_layers, bs, hidden_size, device=device)
    memory = torch.zeros(num_layers, bs, hidden_size, device=device)

    # Disable gradient calculation (we are only doing inference)
    with torch.no_grad():
        # Forward pass through the model
        pred, hidden, memory = model(text_tokens, hidden, memory)

        # Get the predicted category (argmax over the output probabilities)
        predicted_category = pred[:, -1, :].argmax(dim=1).item()

    return predicted_category

# Example usage
text_input = "ChatGPT's AI glitches slows down work"
predicted_category = predict_category(text_input, lstm_classifier, text_tokenizer, text_tranform, device)
print(f"Predicted category: {predicted_category}")
```

```python
import re
import spacy
import torch
import torch.nn as nn
from pdfminer.high_level import extract_text
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator

# Load spaCy model for text processing
nlp = spacy.load("en_core_web_sm")

# Extract text from the PDF file
text = extract_text("problem_sentences.pdf")
cleaned_text = re.sub(r'\s+', ' ', text)

# Process text with spaCy to split into sentences
doc = nlp(cleaned_text)
sentences = [sent.text.strip() for sent in doc.sents]

# Clean and lemmatize sentences
processed_sentences = []
for sentence in sentences:
    doc_sentence = nlp(sentence)
    filtered_sentence = [token.lemma_ for token in doc_sentence if not token.is_stop and not token.is_punct]
    processed_sentences.append(" ".join(filtered_sentence))

# Define a basic tokenizer and build a vocabulary
text_tokenizer = get_tokenizer("basic_english")

# Sample data for building the vocabulary (replace with your own data if available)
sample_data = ["sample text", "example data"]

# Function to yield tokens for building vocabulary
def yield_tokens(data_iter):
    for text in data_iter:
        yield text_tokenizer(text)

# Build vocabulary
vocab = build_vocab_from_iterator(yield_tokens(sample_data), specials=["<unk>"])
vocab.set_default_index(vocab["<unk>"])

# Transformation function to convert tokenized text to tensor format
def text_transform(text_tokens):
    return torch.tensor([vocab[token] for token in text_tokens], dtype=torch.long)

# Define LSTM Model for Classification
class LSTMClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=2):
        super(LSTMClassifier, self).__init__()
        self.embedding = nn.Embedding(input_size, hidden_size)  # Embedding layer
        self.lstm = nn.LSTM(hidden_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x, hidden, memory):
        x = self.embedding(x)  # Pass input through embedding layer
        lstm_out, (hidden, memory) = self.lstm(x, (hidden, memory))  # Pass through LSTM
        out = self.fc(lstm_out[:, -1, :])  # Only take the output from the last time step
        return out, hidden, memory

# Model parameters (replace these with your actual model's parameters)
input_size = len(vocab)  # Size of vocabulary (for embedding layer)
hidden_size = 128
output_size = 4  # Example: 4 categories (World, Sports, Business, Sci/Tech)
num_layers = 2  # Example: Number of layers in your LSTM

# Load your LSTM model and move to the correct device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
lstm_classifier = LSTMClassifier(input_size, hidden_size, output_size, num_layers).to(device)

# Label mapping: Update according to your model's output
label_mapping = {
    0: "World",
    1: "Sports",
    2: "Business",
    3: "Sci/Tech"
}

# Prediction function
def predict_category(text, model, text_tokenizer, text_transform, device):
    model.eval()

    # Tokenize the text (this should be a single string)
    text_tokens = text_tokenizer(text)

    # Transform the tokenized text to tensor format
    text_tensor = text_transform(text_tokens).unsqueeze(0).to(device)  # Add batch dimension (1, sequence_length)

    # Initialize hidden and memory states for LSTM (for batch size of 1)
    bs = 1  # Batch size of 1 for a single input
    hidden = torch.zeros(num_layers, bs, hidden_size, device=device)  # Fixed hidden shape
    memory = torch.zeros(num_layers, bs, hidden_size, device=device)  # Fixed memory shape

    with torch.no_grad():
        # Get predictions from the model
        pred, hidden, memory = model(text_tensor, hidden, memory)

        # Get the predicted category (argmax over the output probabilities)
        predicted_category = pred.argmax(dim=1).item()

    return predicted_category

# Classify each sentence and save category predictions
categorized_sentences = []
for i, sentence in enumerate(processed_sentences):
    category = predict_category(sentence, lstm_classifier, text_tokenizer, text_transform, device)
    categorized_sentences.append((sentences[i], label_mapping[category]))  # Map predicted category to domain

# Print classified sentences
for sentence, category in categorized_sentences:
    print(f"Sentence: {sentence}\nPredicted Domain: {category}\n")

# Save categorized sentences to a PDF (if desired)
pdf_filename = "categorized_sentences.pdf"
c = canvas.Canvas(pdf_filename, pagesize=letter)
width, height = letter

y_position = height - 40  # Starting y position for text
for sentence, category in categorized_sentences:
    if y_position < 40:  # Start a new page if needed
        c.showPage()
        y_position = height - 40
    text = f"Sentence: {sentence}\nCategory: {category}"
    c.drawString(40, y_position, text)
    y_position -= 40  # Move down for the next entry

c.save()
print(f"Categorized sentences saved in '{pdf_filename}'")
```
